```python
# All these methods should be wrapped in user defined fucntions for resuability and exception handling

# A function to handle errors and make code reusable
def getTitle(url):

    """

    The function takes a url, opens it, and reuturns the title of the webpage
    The url is upened with html parser and checks are built for HTTPError AttributeError, and URLError

    arg
        url: The url to be opened
    ret: Title of the webpage

    """

    # Try opening a url
    try:
        html = urlopen(url)
    
    # HTTPError is handled first
    except HTTPError as httpErrorMessage:
        print(httpErrorMessage)
        print("HTTP ERROR MESSAGE")
        return None

    # URLError comes next
    except URLError as urlErrorMessage:
        print(urlErrorMessage)
        print("URL ERROR")
        return None
    
    # If the url is opened successfully, scrape the webpage 
    try:
        
        # Using HTML parser to scrape
        beautifulSoup = BeautifulSoup(html.read(), "html.parser")

        # Get the title of the webpage
        title = beautifulSoup.body.h1

    # Check for attribute error
    except AttributeError as attributeError:

        # Print the error
        print(attributeError)
        print("ATTRIBUTE ERROR")
        return None

    # Return the title
    return title
```

```python
pages = set()
random.seed(datetime.datetime.now())

def getInteralLinks(beautifulSoup, includedUrl):
    includedUrl = "{}://{}".format(urlparse(includedUrl).scheme, urlparse(includedUrl).netloc)
    internalLinks = []
    for link in beautifulSoup.find_all("a", href=re.compile("^(/|.*"+includedUrl+')')):
        if link.attrs["href"] is not None:
            if link.attrs["href"] not in internalLinks:
                if(link.attrs["href"].startswith("/")):
                    internalLinks.append(includedUrl+link.attrs["href"])
                else:
                    internalLinks.append(link.attrs["href"])
    return internalLinks
```

```python
# Retries a list o all external links found on a page

def getExternalLinks(beautufulSoup, excludeUrl):
    
    externalLinks = []
    
    # Find all links that start with "http" that do not contain the current url

    for link in beautifulSoup.find_all("a", href=re.compile("^(http|www)((?!"+excludeUrl+").)*$")):
        if link.attrs["href"] is not None:
            if link.attrs["href"] not in externalLinks:
                externalLinks.append(link.attrs["href"])
    return externalLinks
```

```python
# Collects a list of all external URLs found on the site

allExternalLinks = set()
allInternalLinks = set()

def getAllExternalLinks(siteUrl):
    html = urlopen(siteUrl)
    domain = "{}://{}".format(urlparse(siteUrl).scheme, urlparse(siteUrl).netloc)
    beautifulSoup = BeautifulSoup(html, "html.parser")
    internalLinks = getInteralLinks(beautifulSoup, domain)
    externalLinks = getExternalLinks(beautifulSoup, domain)

    for link in externalLinks:
        if link not in allExternalLinks:
            allExternalLinks.add(link)
            print(link)
    for link in internalLinks:
        if link not in allInternalLinks:
            allInternalLinks.add(link)
            getExternalLinks(link)
```

##### Crawling sites through search

Understanding and leveraging the following conventions is useful in crawling sites through search.

1. Sites often retrieve a list of search resutls for a particular topic by passing that topic as a string through a parameter in teh URL.
2. Search results are presented as an easily identifiable list of links, usually with a convenient tag like <span class = "result">
3. Weather the search resutl conatains relative links or absolute links can be stored as a property of the website object.

While looking for multiple topics in numereous websites, attaintion is to be paid on the load distribution on the servers.

```python
 class Content:
    """
    Common base class for all articles/pages or any type of content on a webpage.

    :topic is the topic on which the article is written
    :url is the link to the webpage on which content is found
    :body is the content 
    :title is the heading of the webpage on which the content is found
    """
    def __init__(self, topic, url, title, body):
        """
         The default constructor for the class.
        """
        # Initialize all the variables of the class
        self.title = title
        self.topic = topic
        self.url = url
        self.body = body

    def print(self):
        """
        Flexible printing function controls output
        """
        print("New article found for topic: {}".format(self.topic))
        print("Title: {}".format(self.title))
        print("Body: \n{}".format(self.body))
        print("URL: {}".format(url))


class Website:
    """
    Contains information about the website structure.

    :name is the name of the website
    :url is the address of the website
    :searchUrl defines where to get search results from anytime a topic is appended
    :resultListing defines the box that holds information about each result
    :resultUrl defines the tag in teh resultListing box that will give the exact url for the result
    :absoluteUrl is a boolen indicating weather the search results are absolute or relative links
    :titleTag
    :bodyTag
    """
    def __init__(self, name, url, searchUrl, resultListing, resultUrl, absoluteUrl
    , titleTag, bodyTag):
        self.name = name
        self.url = url
        self.searchUrl = searchUrl
        self.resultListing = resultListing
        self.resultUrl = resultUrl
        self.absoluteUrl = absoluteUrl
        self.titleTag = titleTag
        self.bodyTag = bodyTag

class Crawler:
    def getPage(self, url):
        try:
            aRequest = requests.get(url)
        except requests.exceptions.RequestException as anException:
            print(anException)
            return None
        return BeautifulSoup(aRequest.text, "html.parser")

    def safeGet(self, pageObject, selector):
        childObject = pageObject.select(selector)
        if childObject is not None and len(childObject) > 0:
            return childObject[0].get_text()
        return ""
        
    def search(self, topic, site):
        """
        Searches a given website for a given topic and records all pages found.
        """
        beautifulSoup = self.getPage(site.searchUrl + topic)
        searchResults = beautifulSoup.select(site.resultListing)
        for result in searchResults:
            url = result.select(site.resultUrl)[0].attrs["href"]
            # Check to see weather it is a relative or absolute url
            if site.absoluteUrl:
                beautifulSoup = self.getPage(url)
            else:
                beautifulSoup = self.getPage(site.url+url)
            if beautifulSoup is None:
                print("Something is wrong with the page or URL, skipping!")
                return
            title = self.safeGet(beautifulSoup, site.titleTag)
            body = self.safeGet(beautifulSoup, site.bodyTag)
            if title != "" and body != "":
                content = Content(topic, title, body, url)
                content.print()

```

```python
class Website:
    def __init__(self, name, url, targetPattern, absoluteUrl, titleTag, bodyTag):
        self.name = name
        self.url = url
        self.targetPattern = targetPattern
        self.absoluteUrl = absoluteUrl
        self.titleTag = titleTag
        self.bodyTag = bodyTag
class Content:
    def __init__(self, url, title, body):
        self.url = url
        self.title = title
        self.body = body

    def print(self):
        print("URL: {}".format(self.url))
        printt("Title: {}".format(self.title))
        print("Body: \n{}".format(self.body))
class Crawler:
    def __init__(self, site):
        self.site = site
        self.visited = []
    def getPage(self, url):
        try:
            aRequest = requests.get(url)
        except requests.exceptions.RequestException:
            return None
        return BeautifulSoup(aRequest.text, "html.parser")
    def safeGet(self, pageObject, selector):
        selectedElements = pageObject.select(selector)
        if selectedElements is not None and len(selectedElements) > 0:
            return "\n".join([element.get_text() for element in selectedElements])
        return ""
    def parse(self, url):
        beautifulSoup = self.getPage(url)
        if beautifulSoup is not None:
            title = self.safeGet(beautifulSoup, self.site.titleTag)
            body = self.safeGet(beautifulSoup, self.site.bodyTag)
            if title != "" and body != "":
                content = Content(url, title, body)
                content.print()
    def crawl(self):
        """
        Get pages from sites home page.
        """
        beautifulSoup = self.getPage(self.site.url)
        targetPages = beautifulSoup.find_all("a", href = re.compile(self.site.targetPattern))
        for targetPage in targetPages:
            targetPage=targetPage.attrs["href"]
            if targetPage not in self.visited:
                self.visited.append(targetPage)
                if not self.site.absoluteUrl:
                    targetPage = "{}{}".format(self.site.url, targetPage)
                self.parse(targetPage)
```
